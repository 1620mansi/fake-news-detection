```python
df = pd.read_csv('/kaggle/input/textdb3/fake_or_real_news.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['label'] = encoder.fit_transform(df['label'])
df.head()
```

```python
x = df["title"]
y = df["label"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000,stop_words='english')
x = cv.fit_transform(df['title']).toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, ComplementNB,GaussianNB
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
cnb = ComplementNB()
```

```python
gnb.fit(xtrain,ytrain)
mnb.fit(xtrain,ytrain)
bnb.fit(xtrain,ytrain)
cnb.fit(xtrain,ytrain)
```

```python
y_pred1 = gnb.predict(xtest)
y_pred2 = mnb.predict(xtest)
y_pred3 = bnb.predict(xtest)
y_pred4 = cnb.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,y_pred1))
print(accuracy_score(ytest,y_pred2))
print(accuracy_score(ytest,y_pred3))
print(accuracy_score(ytest,y_pred4))
```

```python
result = bnb.predict(cv.transform(["Aamir Khan's marriage with Kiran Rao and Reena Dutta each lasted 16 years"]).toarray())[0]
if result==0:
    print('FAKE')
else:
    print('REAL')
```
